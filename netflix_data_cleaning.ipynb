```python
# ==========================================
# TASK 1: DATA CLEANING & PREPROCESSING
# Netflix Movies and TV Shows Dataset
# ==========================================

import pandas as pd
import numpy as np

# To upload the file, run this cell and then select 'netflix_titles.csv' from your local machine.
from google.colab import files
import os

# Check if the file already exists to avoid re-uploading
if not os.path.exists("netflix_titles.csv"):
    uploaded = files.upload()
    for fn in uploaded.keys():
        print(f'User uploaded file "{fn}"')

# Load Dataset
df = pd.read_csv("netflix_titles.csv")

print("Dataset Shape Before Cleaning:", df.shape)

# ------------------------------------------
# 1. Check Missing Values
# ------------------------------------------
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

# ------------------------------------------
# 2. Handle Missing Values
# ------------------------------------------
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not Available')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')

# Remove rows where date_added is missing
df = df.dropna(subset=['date_added'])

# ------------------------------------------
# 3. Remove Duplicate Rows
# ------------------------------------------
duplicates_before = df.duplicated().sum()
print("\nDuplicate Rows Found:", duplicates_before)

df = df.drop_duplicates()

duplicates_after = df.duplicated().sum()
print("Duplicate Rows After Removal:", duplicates_after)

# ------------------------------------------
# 4. Clean Column Names
# ------------------------------------------
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
)

# ------------------------------------------
# 5. Standardize Text Columns
# ------------------------------------------
text_columns = ['type', 'country', 'rating']

for col in text_columns:
    df[col] = df[col].astype(str).str.strip()

# ------------------------------------------
# 6. Convert Date Format
# ------------------------------------------
df['date_added'] = pd.to_datetime(
    df['date_added'],
    errors='coerce'
)

# Format as DD-MM-YYYY
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')

# ------------------------------------------
# 7. Fix Data Types
# ------------------------------------------
df['release_year'] = df['release_year'].astype(int)

# ------------------------------------------
# 8. Final Validation
# ------------------------------------------
print("\nDataset Information After Cleaning:")
print(df.info())

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

print("\nDataset Shape After Cleaning:")
print(df.shape)

# ------------------------------------------
# 9. Save Cleaned Dataset
# ------------------------------------------
df.to_csv("netflix_titles_cleaned.csv", index=False)

print("\nCleaning Completed Successfully!")
print("Cleaned file saved as:")
print("netflix_titles_cleaned.csv")
```
